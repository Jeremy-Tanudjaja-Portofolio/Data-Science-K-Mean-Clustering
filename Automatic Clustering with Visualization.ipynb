```python
# Imports
import pandas as pd
import numpy as np
```

## Data Preprocessing

#### At this Stage we are going to grab our data in a csv format, then we are going to do a bit of pre processing before we input it into our K-Means Algorithm

```python
players = pd.read_csv("players_22.csv")
```

```python
# See what the column names are
players.head()
```

```python
# Manually grabs the column name
features = ["overall", "potential", "wage_eur", "value_eur", "age"]
features
```

```python
# Drop Null values
players = players.dropna(subset=features)
```

```python
# Get all the features
data = players[features].copy()
```

```python
data.sample(10)
```

# TODO's

### High level
1. Scale data to standardize values
2. Initialize random centroids
3. Get labels for each data point
4. Create new centroids
5. Plot the centroids
6. Repeat 3-5 until the centroids stop changing

```python
# Data Normalization to 1-11 using min max normalization
data = ((data - data.min()) / (data.max() - data.min())) * 10 + 1
```

```python
data.describe()
```

```python
data.head()
```

```python
# Function to get Random Centroids
def random_centroids(data, k):
    centroids = []
    for i in range(k):
        centroid = data.apply(lambda x: float(x.sample()))
        centroids.append(centroid)
    return pd.concat(centroids, axis=1)
```

```python
# Get 5 Random Centroids
centroids = random_centroids(data, 5)
centroids
```

```python
# Get the Cluster Member
def get_labels(data, centroids):
    distances = centroids.apply(lambda x: np.sqrt(((data - x) ** 2).sum(axis=1)))
    return distances.idxmin(axis=1)
```

```python
labels = get_labels(data, centroids)
```

```python
# How many members in each cluster?
labels.value_counts()
```

```python
# Update the centroid coordinates
def new_centroids(data, labels, k):
    centroids = data.groupby(labels).apply(lambda x: np.exp(np.log(x).mean())).T
    return centroids
```

## Visualization using PCA

```python
# Import for visualization using PCA
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from IPython.display import clear_output
```

```python
def plot_clusters(data, labels, centroids, iteration):
    pca = PCA(n_components=2)
    data_2d = pca.fit_transform(data)
    centroids_2d = pca.transform(centroids.T)
    clear_output(wait=True)
    plt.title(f'Iteration {iteration}')
    plt.scatter(x=data_2d[:,0], y=data_2d[:,1], c=labels)
    plt.scatter(x=centroids_2d[:,0], y=centroids_2d[:,1])
    plt.show()
```

```python
# Loop Logic

max_iterations = 100
centroid_count = 3

centroids = random_centroids(data, centroid_count)
old_centroids = pd.DataFrame()
iteration = 1

while iteration < max_iterations and not centroids.equals(old_centroids):
    old_centroids = centroids

    labels = get_labels(data, centroids)
    centroids = new_centroids(data, labels, centroid_count)
    plot_clusters(data, labels, centroids, iteration)
    iteration += 1
```
